# Định nghĩa hàm
Mn thu gọn cho dễ nhìn

```python
csv_folder = os.path.join(db_folder, 'map-keyframes')
```

```python
# KIEM TRA SO LUONG CAC FILE FRAME CHO THU MUC VIDEO CUA TUNG GROUP
def check_folders(group):
    group_folder = os.path.join(db_folder, f'KeyFrames_{group}')
    paths = sorted(glob.glob(f'{group_folder}/L*'))
    txts = glob.glob(f'{group_folder}/*.txt')
    # Lấy các thư mục
    paths = list(filter(lambda i: i not in txts, paths))
    counts = []
    for path in paths:
        counts.append(len(glob.glob(f'{path}/*.jpg')))
    return ([os.path.split(path)[-1] for path in paths], counts)

def check_txts(group):
    group_folder = os.path.join(db_folder, f'KeyFrames_{group}')
    paths = sorted(glob.glob(f'{group_folder}/*.txt'))
    return [os.path.split(path)[-1] for path in paths]

def check_csvs(group):
    paths = sorted(glob.glob(f'{csv_folder}/{group}_*.csv'))
    counts = []
    for path in paths:
        df = pd.read_csv(path)
        counts.append(df.shape[0])
    return ([os.path.split(path)[-1] for path in paths], counts)

def is_same(folder, csvname, txtname):
    csvname = csvname.replace('.csv','')
    txtname = txtname.replace('.txt','')
    if folder != csvname or folder != txtname or csvname != txtname:
        return False
    return True
```

```python
def print_row(name_list):
    print(f'Count: {len(name_list)}')
    for name in name_list:
        print(name)

def print_3row(name_tuple, csv_tuple, txt_list, min_count=600):
    folders, f_counts = name_tuple
    csv_list, c_counts = csv_tuple
    print(f'Count folder: {len(folders)}')
    print(f'Count csv: {len(csv_list)}')
    print(f'Count txt: {len(txt_list)}')
    for folder, f_count, cs, c_count, tx in zip(folders, f_counts, csv_list, c_counts, txt_list):
        if is_same(folder, cs, tx) and f_count == c_count:
            if f_count > min_count:
                print(f'{folder}__{cs}__{tx}__({f_count})__({c_count})')
            else:
                print(f'NumErr=>{folder}__{cs}__{tx}__({f_count})__({c_count})')
        else:
            print(f'NotMatchErr=>{folder}__{cs}__{tx}__({f_count})__({c_count})!!!')
```

```python
%%time
# ===== NHẬP TÊN GROUP ====
# Một group chứa nhiều video hay folder video (ví dụ: 'L01')
group = 'L16'
# ========================

print_3row(check_folders(group), check_csvs(group), check_txts(group))

# CÁC THÔNG TIN HIỂN THỊ
# Count folder: số lượng folder video trong KeyFrames_{group}
# Count csv: số lượng file csv của group
# Count txt: số lượng file txt của group
# <Tên folder frame>__<Tên csv>__<Tên txt>__(<slg ảnh trong folder>)__(<slg dòng csv>)

# NotMatchErr xảy ra do các (tên không giống nhau) hoặc (slg không giống nhau)
# NumErr xảy ra do slg của cả ảnh và dòng csv dưới 600
# Nếu cả một dòng NotMatchErr thì có thể do bị thiếu mất file, folder nào đó
```
